```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
samps = 5000
```

```python
X = np.random.randn(2,samps)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(X[0],X[1], alpha=0.3, s=2);
ax.set_aspect(1)
ax.set_xticks([-2,0,2])
ax.set_yticks([-2,0,2]);
```

```python
X2 = X.copy()
X2[1] = X[1]+0.5*X[0]
```

```python
X2svd = np.linalg.svd(X2@X2.T/samps)
```

```python
Y = X2[1]-0.3*X2[0] + 0.2*np.random.randn(1,samps)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(X2[0],X2[1], alpha=0.2, s=2, color='black');
ax.set_aspect(1)
# ax.set_ylim(-4,4);
ax.set_xlabel('$x_1$',fontsize=17);
ax.set_ylabel('$x_2$',fontsize=17);
a1,b1 = -X2svd[0][0]
line1 = ax.plot([-10*a1,10*a1],[-10*b1,10*b1])[0];
a2,b2 = X2svd[0][1]
line2 = ax.plot([-10*a2,10*a2],[-10*b2,10*b2])[0];
a3,b3 = -0.2,1
line3 = ax.plot([-10*a3,10*a3],[-10*b3,10*b3])[0];
ax.set_ylim(-4.3,4.3);
ax.set_xlim(-3.6,3.6);
ax.set_yticks([-3,0,3]);
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
corr = (Y[0]*(a3*X2[0]+b3*X2[1])).sum()/np.sqrt((Y[0]**2).sum()*((a3*X2[0]+b3*X2[1])**2).sum())
ax.scatter(a3*X2[0]+b3*X2[1],Y[0], alpha=0.3, s=2, color = line3.get_color(), label = 'correlation = {:.2f}'.format(corr));
ax.set_aspect(1)
ax.spines['bottom'].set_color(line3.get_color());
ax.tick_params(axis='x', colors=line3.get_color())
ax.set_xlabel('$p_3$',fontsize=17, color=line3.get_color());
ax.set_ylabel('$y$',fontsize=17);
ax.set_yticks([-4,-2,0,2,4]);
# plt.legend(loc=1, fontsize=14);

leg = ax.legend(handlelength=0, handletextpad=0, fancybox=True,loc=4, fontsize=14);
for item in leg.legendHandles:
    item.set_visible(False)
```
